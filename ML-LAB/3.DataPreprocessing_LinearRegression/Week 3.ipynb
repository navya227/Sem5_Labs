# Q2

```python
df2 = pd.read_csv('Lab3.csv')
```

```python
x = df2['time'].values
y = df2['score'].values
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

num = np.sum((x - x_mean) * (y - y_mean))
den = np.sum((x - x_mean)**2)
b1 = num / den
b0 = y_mean - b1 * x_mean

y_pred = b0 + b1 * x

rmse = np.sqrt(np.mean((y - y_pred)**2))
b0, b1, rmse, y_pred
```

```python
plt.scatter(x, y, color='red', label='Data Points')

plt.plot(x, y_pred, color='blue', label='Regression Line')

plt.xlabel('Study Time')
plt.ylabel('Score')
plt.title('Linear Regression')
plt.legend()

plt.show()
```

```python
n = len(x)
sum_x = np.sum(x)
sum_x2 = np.sum(x**2)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)

A = np.array([[n, sum_x], [sum_x, sum_x2]])
B = np.array([sum_y, sum_xy])

A_inv = np.linalg.inv(A)

coef= A_inv @ B
B0, B1= coef

B0, B1
```

```python
test = 10
pred = b0 + b1 * test
pred2 = B0 + B1 * test

print(pred)
print(pred2)
```
