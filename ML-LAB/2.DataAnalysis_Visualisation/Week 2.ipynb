# Q1

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
x = [0,20,40,60,80,100]
y = [2*xi for xi in x] 

ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Plot')

plt.show()
```

# Q2

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x, y, label='y = 2x')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Plot on ax1')
ax1.legend()

ax2.plot(x, y, label='y = 2x', color='orange')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Plot on ax2')
ax2.legend()

plt.show()
```

# Q3

```python
import pandas as pd

data = pd.read_csv('company_sales_data.csv')

plt.plot(data['month_number'], data['total_profit'])

plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit per Month')

plt.show()
```

# Q4

```python
plt.plot(
    data['month_number'], 
    data['total_profit'], 
    linestyle=':',      
    color='red',        
    linewidth=3,        
    marker='o',         
    markerfacecolor='red',
    label='Total Profit'
)

plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.title('Total Profit per Month')

plt.legend(loc='lower right')

plt.show()
```

# A1

```python

months = data['month_number']
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

plt.figure(figsize=(12, 8))

for product in products:
    plt.plot(months, data[product], marker='o')

plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Monthly Sales Data for Each Product')

plt.show()
```

# A2

```python
total_units_sold = data[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].sum()

products = total_units_sold.index
units_sold = total_units_sold.values

plt.figure(figsize=(10, 7))
plt.pie(units_sold, labels=products, autopct='%1.1f%%', startangle=140)
plt.title('Percentage')
plt.show()
```
